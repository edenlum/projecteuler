<p>The sequence of triangle numbers is generated by adding the natural numbers. So the $7$<sup>th</sup> triangle number would be $1 + 2 + 3 + 4 + 5 + 6 + 7 = 28$. The first ten terms would be:
$$1, 3, 6, 10, 15, 21, 28, 36, 45, 55, \dots$$</p>
<p>Let us list the factors of the first seven triangle numbers:</p>
\begin{align}
\mathbf 1 &amp;\colon 1\\
\mathbf 3 &amp;\colon 1,3\\
\mathbf 6 &amp;\colon 1,2,3,6\\
\mathbf{10} &amp;\colon 1,2,5,10\\
\mathbf{15} &amp;\colon 1,3,5,15\\
\mathbf{21} &amp;\colon 1,3,7,21\\
\mathbf{28} &amp;\colon 1,2,4,7,14,28
\end{align}
<p>We can see that $28$ is the first triangle number to have over five divisors.</p>
<p>What is the value of the first triangle number to have over five hundred divisors?</p>


```python
# the n'th triangle number is equal to n(n+1)/2
# check with 7: 7*8/2 = 7*4=28
# therefore the prime factors of the nth triangle number are the prime factors of n and the prime factors of (n+1)/2
# the number of divisors are given from the prime factors by considering the power of each factor and multiplying the powers + 1
# for example for 28 = 2*2*7 = 2^2 * 7^1 -> (2+1)*(1+1) = 6
```

```python
def prime_factors(n):
    factors = {}
    i = 2
    while i * i <= n:
        count = 0
        while n % i == 0:
            count += 1
            n = n // i
        if count > 0:
            factors[i] = count
        i += 1
    if n > 1:
        factors[n] = 1
    return factors

def prod(lst):
    p = 1
    for i in lst:
        p *= i
    return p

n = 1
while True:
    triangle = n*(n+1)//2
    prime_factors_triangle = prime_factors(n)
    for prime, power in prime_factors((n+1)//2).items():
        if prime in prime_factors_triangle:
            prime_factors_triangle[prime] = prime_factors_triangle[prime] + power
        else:
            prime_factors_triangle[prime] = power
    print(triangle, prime_factors_triangle)
    num_divisors = prod([power+1 for prime, power in prime_factors_triangle.items()])
    print(num_divisors)
    if num_divisors > 500:
        break
    n += 1
```
